```python
import cv2
import easyocr
```

```python
import os
import numpy as np
```

```python
data_train=r"C:\Users\trung\Desktop\CV_Projects\Train"
data_test=r"C:\Users\trung\Desktop\CV_Projects\Test"

```

```python
X_train=[]
y_train=[]

X_test=[]
y_test=[]
for dir in os.listdir(data_train):
    directory=os.path.join(data_train,dir)
    for file in os.listdir(directory):
        file=os.path.join(directory,file)
        image=cv2.imread(file)
        image=cv2.resize(image,(20,20))
        image=np.array(image).flatten()
        X_train.append(image)
        y_train.append(dir)

for dir in os.listdir(data_test):
    directory=os.path.join(data_test,dir)
    for file in os.listdir(directory):
        file=os.path.join(directory,file)
        image=cv2.imread(file)
        image=cv2.resize(image,(20,20))
        image=np.array(image).flatten()
        X_test.append(image)
        y_test.append(dir)
```

```python
from sklearn import svm
```

```python
svm_default=svm.SVC(kernel='linear').fit(X_train,y_train)
```

```python
svm_default.score(X_test,y_test)
```

```python
import pickle
```

```python
pickle.dump(svm_default, open("license_plate_model_svm.yml", 'wb'))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
softmax_default=LogisticRegression(multi_class="multinomial",max_iter=250).fit(X_train,y_train)
```

```python
softmax_default.score(X_test,y_test)
```

```python
pickle.dump(softmax_default, open("license_plate_model_softmax.yml", 'wb'))
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split
```

```python
svm=pickle.load(open("license_plate_model_svm.yml","rb"))
softmax=pickle.load(open("license_plate_model_softmax.yml","rb"))
```

```python
pred_svm=svm.predict(X_test)
pred_softmax= softmax.predict(X_test)
```

```python
print("acc: svm ",svm.score(X_test,y_test)," softmax ", softmax.score(X_test,y_test))
print("f1_score: svm ",f1_score(pred_svm,y_test,average="macro")," softmax ", f1_score(pred_softmax,y_test,average="macro"))
print("precission: svm ",precision_score(pred_svm,y_test,average="macro")," softmax ", precision_score(pred_softmax,y_test,average="macro"))
print("Recall: svm ",recall_score(pred_svm,y_test,average="macro")," softmax ", recall_score(pred_softmax,y_test,average="macro"))
```
